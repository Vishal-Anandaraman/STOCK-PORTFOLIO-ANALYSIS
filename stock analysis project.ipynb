```python
tickers = ['TCS.NS', 'SUNPHARMA.NS', 'ADANIGREEN.NS', 'ONGC.NS']
start_date = '2020-01-01'
end_date = '2024-12-31'
```

```python
#Sharpe Ratio (Annualized) ( risk free rate as 4% per year, 252 trading days)
```

```python
risk_free_rate = 0.04 
trading_days = 252
```

```python
annual_return = daily_returns.mean() * trading_days
annual_volatility = daily_returns.std() * (trading_days ** 0.5)
```

```python
sharpe_ratio = (annual_return - risk_free_rate) / annual_volatility
```

```python
print("Sharpe Ratios (Risk-free rate = 4%):")
print(sharpe_ratio.sort_values(ascending=False))
sharpe_ratio.sort_values().plot(kind='barh', figsize=(8,5), title='Sharpe Ratios by Stock')
plt.xlabel("Sharpe Ratio")
plt.grid(True)
plt.show()
```
